```python
#DATA PERBANDINGA PENONTON NETFLIX MOVIE DAN TV SHOW
dataset = "netflix_titles.csv"
df = pd.read_csv(dataset)
df.head(10)
```

PIE CHART 

```python
from collections import Counter
print(Counter(df['type']))
```

```python
title_type = df.groupby('type').agg('count')
print(title_type)
```

```python
type_labels = title_type.show_id.sort_values().index 
type_counts = title_type.show_id.sort_values()
```

```python
type_labels
```

```python
type_counts
```

```python
plt.figure(1, figsize=(20,10)) 
the_grid = GridSpec(2, 2)
```

```python
cmap = plt.get_cmap('Spectral')
colors = [cmap(i) for i in np.linspace(0, 1, 8)]
```

```python
plt.subplot(the_grid[0, 1], aspect=1, title='Types of Netflix Titles')
type_show_ids = plt.pie(
    type_counts,
    labels=type_labels,
    autopct='%1.1f%%',
    colors=colors,
    shadow=True,
)
plt.show()
```

Correlation dan Causation

```python
# import library scipy
from scipy import stats

# memuat data harga rumah di amerika serikat
path ='house price.csv'
df = pd.read_csv(path)
df.head()
```

```python
# menghitung koefisien korelasi pearson dari 'wheels-base' & 'price'
pearson_coef, p_value = stats.pearsonr(df['price'], df['bathrooms'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```

```python
# visualisasi scatter plot menggunakan regplot untuk mengetahui hubungan korelasi variabel 'engine-size' & 'price'
sns.regplot(x="bathrooms", y="price", data=df)
plt.ylim(0,)
```

```python
# mengetahui korelasi 'bathrooms' & 'price'
df[["bathrooms", "price"]].corr()
```

Grouping

```python
# menampilkan nilai unik data series
df['statezip'].unique()
```

```python
# membuat dataframe baru dengna kolom 'statezip','city', dan 'price'
df_group_one = df[['statezip','city','price']]
```

```python
# melakukan grouping
df_group_one = df_group_one.groupby(['statezip'],as_index=False).mean()
df_group_one
```

```python
# melakukan grouping
df_gptest = df[['statezip','city','price']]
grouped_test1 = df_gptest.groupby(['statezip','city'],as_index=False).mean()
grouped_test1
```

```python
# melakukan pivot data 
grouped_pivot = grouped_test1.pivot(index='statezip',columns='city')
grouped_pivot
```

```python
# melakukan pivot data dengan handle missing value dengan nilai 0
grouped_pivot = grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot
```

```python
# melakukan grouping dari  "city" berdasarkan rata-rata "harga"
df_gptest2 = df[['city','price']]
grouped_test_bodystyle = df_gptest2.groupby(['city'],as_index= False).mean()
grouped_test_bodystyle
```

```python
#visualisasi heatmap
plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

# nama label
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

# pindahkan ticks and labels ke tenah
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

# masukan  labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

# rotasi label
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```
